```python
#Importing the needed libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel(r'C:/Users/Dell/Desktop/Assignments/Clustering/EastWestAirlines.xlsx',sheet_name='data')
data.head()
```

```python
#Dropping unnecessary columns
df = data.drop(['ID#','cc1_miles','cc2_miles','cc3_miles','Award?'],axis=1)
```

```python
array = data.values
```

```python
array
```

```python
stdscaler = StandardScaler().fit(array)
X = stdscaler.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=6)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
clusters = pd.DataFrame(dbscan.labels_,columns=['Cluster_ID'])
clusters.head()
```

```python
data1 = pd.concat([df,clusters],axis=1)
data1
```

```python
data1.groupby('Cluster_ID').count()
```

```python
data1.groupby('Cluster_ID').agg(['mean']).reset_index()
```

* Qual_miles with the clusters 3 to 7 are zero, showing little similarity
