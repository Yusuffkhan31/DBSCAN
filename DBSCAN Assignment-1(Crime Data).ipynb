```python
#Importing the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/Users/Dell/Desktop/Assignments/Clustering/crime_data.csv')
data.head()
```

```python
data1=data.rename(columns={'Unnamed: 0':'State'})
data1.head()
```

```python
data1.drop(['State'], axis=1,inplace=True)
data1.head()
```

```python
array = data1.values
```

```python
array
```

```python
stdscaler = StandardScaler().fit(array)
STD = stdscaler.transform(array)
```

```python
STD
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=3)
dbscan.fit(STD)
```

```python
dbscan.labels_
```

```python
cluster = pd.DataFrame(dbscan.labels_,columns=['Clusters'])
cluster
```

```python
df = pd.concat([data1,cluster],axis=1)
```

```python
df.head()
```

```python
df.groupby('Clusters').count()
```

# Inferences
* With the change in the value of minimum sample, the number of clusters formed change considerably
* Clusters one and two cover a very small section of the dataset
